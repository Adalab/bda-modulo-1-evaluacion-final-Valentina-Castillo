```python
#Todo este codigo es la plantilla para crear una tienda online

class OnlineStore:

   def __init__(self):  # constructor. se ejecuta automaticamente al crear una tienda nueva.
     self.inventory = [] #lista vacia para guardar productos
     self.total_sales = 0 #contador de ventas
     self.clients = {} #diccionario vacio para guardar clientes (BONUS)

#Add products to the inventory
   def add_product(self, name, price, quantity):
        for product in self.inventory: #Recorre cada producto del inventario
           if product ["name"] == name:    #para ver si el producto ya existe
              product ["quantity"] += quantity #sumar la cantidad nueva con la que ya hay en stock
              break #sale del bucle si se encontro el producto
        else:
            new_product = {"name": name, "price": price, "quantity": quantity}
            self.inventory.append(new_product) #else del for. En caso de no existir se crea un nuevo diccionario con los datos nuevos y lo añade al inventario
#see inventory : Muestra TODOS los productos con sus detalles
   def view_inventory(self):
       for product in self.inventory: #itera a traves del inventario con el blucle for
          print(f"name = {product["name"]}, price = {product["price"]}, quantity = {product["quantity"]}")

#built store : crea una tienda nueva y añade productos
Clothing_store = OnlineStore()
Clothing_store.add_product("Shirt", 20,  50)
Clothing_store.add_product("Pants", 30,  30 )
Clothing_store.add_product("Shoes", 50,  40 )
Clothing_store.add_product("Shirt", 20,  50 )

 #serch for product: Busca SOLO el producto especifico y muestra sus datos-
def search_product(self,name):
        for product in self.inventory: #Recorre el inventario
           if product ["name"] == name:
              print(f"name = {product["name"]}, price = {product["price"]}, quantity = {product["quantity"]}")
              break #si encuentra el producto lo printea y sale del bucle.
        else:
              print("product not found") #no encuentra el producto en el bucle
#update stock: cambia la cantidad de un producto
def update_stock(self, name, quantity):
        for product in self.inventory:
          if product ["name"] == name: #si el producto esta en el inventario se suma la cantidad de este a la que haya en inventario
             product ["quantity"] += quantity
             print("stock updated")
             break
        else: 
            print("product not found")
# delete product: eliminia completamente un producto del inventario
def delete_product(self, name):
        for product in self.inventory:
           if product ["name"] == name:
              self.inventory.remove(product) #elimina el producto de la lista
              print("product removed")
              break
        else: 
              print("product not in inventory")
         
#calculate value: Calcula cuanto vale todo el inventario
def calculate_inventory_value(self):
        total = 0 #contador empieza en 0
        for product in self.inventory: #recorre cada producto
           total += product["price"] * product["quantity"] #suma (precio * cantidad)
        print(f"Total inventory value: {total}")
        




#BONUS

#1 Registramos nuevos clientes al diccionario. 
# Usa el nombre como clave y guarda un diccionario con su email y una lista vacía para el historial de compras.
def add_client(self,name,email):
        self.clients[name] = {"email": email, "purchases": []}

#2 Muestra todos clientes registrados
def view_clients(self):
        for client, info in self.clients.items(): #Recorre el dicc de clientes. devuelve pares (clave/valor)
         print(f"name: {client}, email: {info["email"]}")  #cliente : clave / email: valor

#3 Realizar una compra
def make_purchase(self):
 cart = {} #diccionario vacio
 total = 0

 client_name = input("Enter your name") #para verificar si el cliente esta registrado

 if client_name not in self.clients:
    print("client not registered")
    return 
      

 while True: #Recorre toda la base de datos
       
    self.view_inventory() #Mostramos el inventario de productos llamandolo

    purchase_name = input("please insert the pruduct you want to purchase") # pedimos al cliente el producto que quiere.
    
        

    for purchase in self.inventory: #bulce para recorrer el inventario
        
           if purchase in self.inventory: #verificar si hay stock suficiente
              purchase_quantity = int(input( "How many do you want to purchase?")) # pedimos la cantidad

              if purchase_quantity <= self.inventory["sotck"]: 
                  self.inventory[purchase_name]["stock"] -= purchase_quantity     #se reduce el stock
         
               #añadir producto al carrito
              cart.append(({"name": purchase_name, "price": self.inventory[purchase]["price"] , "quantity": purchase_quantity}))

               #calcular total
              total += self.inventory[purchase_name]["price"] * purchase_quantity
       
           else: 
             print("no stock available")

 else: 
   print("product entered does not exist")
      
             
```
